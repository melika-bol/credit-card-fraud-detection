```python
import os
import pandas as pd
csv_path = os.path.join(path, "creditcard.csv")
df = pd.read_csv(csv_path)
```

```python
fraud = df[df['Class'] == 1]
valid = df[df['Class'] == 0]
outlierfraction = len(fraud)/len(valid)  #It calculates the ratio of fraud cases to valid cases to measure how imbalanced the dataset is.
print(outlierfraction)
print('Fraud Cases: {}'.format(len(fraud)))
print('Valid Transactions: {}'.format(len(valid)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
#correlation between features
corr = df.corr()
fig = plt.figure(figsize = (8,8))
sns.heatmap(corr, cmap='coolwarm', annot=False, fmt=".2f")
plt.show()
```

```python
#To understand which features are important in fraud detection.

features = ['Time','V2','V5','V7','V12','V20','Amount']
for feature in features:
  plt.figure(figsize=(6,3))
  sns.kdeplot(data=valid, x=feature, label='Valid', fill=True)
  sns.kdeplot(data=fraud, x=feature, label='Fraud', fill=True)
  plt.title(f'Distribution of {feature} by Class')
  plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
#Do fraudulent transactions usually have certain values?
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

```python
#Do frauds occur more often at certain times of the day?

plt.figure(figsize=(8, 4))
sns.histplot(fraud['Time'], bins=100, color='red', label='Fraud', alpha=0.6)
sns.histplot(valid['Time'], bins=100, color='green', label='Non-Fraud', alpha=0.6)
plt.legend()
plt.title('Time Distribution by Class')
plt.xlabel('Time (seconds since first transaction)')
plt.show()
```

```python
#Sometimes the features that are related only in fraudulent transactions are hidden in the whole data.

fraud_corr = fraud.corr()
plt.figure(figsize=(10,8))
sns.heatmap(fraud_corr, cmap='coolwarm', annot=False, fmt=".2f")
plt.title("Correlation Matrix - Fraud Only")
plt.show()
```

```python
#Which features in most frauds have very large or negative values

sns.boxplot(data=fraud[['V10', 'V12', 'V14']])
plt.title('Boxplot of Important Features in Fraudulent Transactions')
plt.show()
```
